## Normalización: eliminar skewness

```python
from sklearn.preprocessing import PowerTransformer
```

```python
# Comprobamos que el df no tiene valores nulos ni negativos

for a in ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper','Delicassen']:
    print(f'Variable {a} has {len(df[df[a] == 0])} zero values')
    print(f'Variable {a} has {len(df[df[a] < 0])} negative values')
```

```python
# Definimos nuestra matriz X con las variables numéricas

X = df.drop(['Channel', 'Region'], axis=1)
```

```python
# Creamos el modelo

pt = PowerTransformer(method='yeo-johnson')
```

```python
# Aplicamos fit_transform a nuestros datos

data = pt.fit_transform(X)
```

```python
# Visualizamos las nuevas distribuciones normalizadas

fig, ax = plt.subplots(3,2, figsize=(20,18))
var = list(range(0,6))

for i in range(3):
    for j in range(2):
        x = var.pop(-1) # Con .pop() extrae el último valor de la lista y nos ahorramos otra iteración
        sns.distplot(data[:,x], ax=ax[i][j])
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

##### NOTA: ESTA PARTE LA TENGO EN CÓDIGO PERO AL HABER APLICADO BOX-COX YA TENGO LOS DATOS NORMALIZADOS.

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['Channel', 'Region'], axis=1)
```

```python
model = StandardScaler()
```

```python
model.fit(X)
```

```python
customers_scale = model.transform(X)
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
from ipywidgets import interact
```

```python
# Hacemos una función para visualizar los resultados en función del número de clusters.

# Esta función podría escalarse de forma que aceptase el df y se le hiciera la transformación

# Por ahora la mantenemos así simplemente para que se aplique a nuestros datos anteriormente calculados

@interact(n=range(2,9))

def KM(n):
    """
    This function prints the result of applying different number of clusters to K-Means to our
    normalized dataset 'customers_scale' by visualizing it through our 2-dimension reduced new
    dataframe 'pc'.
    
    Input: nr. clusters
    """
    
    test = pd.DataFrame(pc[:,:2], columns=['t0', 't1'])
    test['lb'] = KMeans(n).fit(customers_scale).labels_
    test['lb'] = test['lb'].apply(lambda x: str(x))
    sns.scatterplot(data=test, x='t0', y='t1', hue='lb')
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Hacemos una función para visualizar los resultados en función de los parámetros.

# Esta función podría escalarse de forma que aceptase el df y se le hiciera la transformación

# Por ahora la mantenemos así simplemente para que se aplique a nuestros datos anteriormente calculados

@interact(eps=np.arange(0.25, 0.8, 0.05), min_samples=range(2,8))

def custom_DBSCAN(eps, min_samples):
    """
    This function prints the result of applying different eps & min_samples values to DBSCAN to our
    normalized dataset 'customers_scale' by visualizing it through our 2-dimension reduced new
    dataframe 'pc'.
    
    Input: eps, min_samples
    """
    
    test = pd.DataFrame(pc[:,:2], columns=['t0', 't1'])
    test['lb'] = DBSCAN(eps=eps, min_samples=min_samples).fit(customers_scale).labels_
    test['lb'] = test['lb'].apply(lambda x: str(x))
    sns.scatterplot(data=test, x='t0', y='t1', hue='lb')
```
